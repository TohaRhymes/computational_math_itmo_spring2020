## Interpolation

```python
import numpy as np
from numpy import genfromtxt
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
from matplotlib.pyplot import figure
import sys
import math
import pandas as pd
from math import factorial as fc
```

#### Input and draw

```python
def in_float(s = 'Введите число', integer = False, check = [False, 0, 0]):
    flag = True
    while flag:
        flag = False
        try:
            if integer:
                val = int(input(s +': '))
            else:
                val = float(input(s +': '))
            if check[0] and (val < check[1] or val > check[2]):
                raise ValueError
        except ValueError:
            flag = True
            if check[0]:
                print(f'Попробуйте снова! Введенное число должно принадлежать интервалу [{check[1]}; {check[2]}]\n')
            else:
                print(f'Попробуйте снова!\n')
    return val

def parse():
    flag = True
    while flag:
        path = input('Введи путь:\n').strip()
        try:
            a = genfromtxt(path, delimiter=',')
            if True in np.isnan(a) or a.shape[0] != 2:
                raise ValueError
            return a   
        except ValueError:
            print('В файле должно быть 2 строчки, в каждой одинаковое количество чисел\n')
        except OSError:
            print('Такого файла нет.\n')
        print('Попробуйте снова!\n')

def input_vals():
    n = in_float(s = 'Введите количество точек', integer = True)
    print()
    a = []
    for i in range(int(n)):
        a.append([in_float('x'), in_float('y')])
        print()
    return np.array(a).transpose()

def to_df(array, f, eps, f_type = 'f'):
    np_arr = np.concatenate((array, [f], [eps]), axis=0)
    return pd.DataFrame(data=np_arr, index=["X", "Y", f_type, 'eps'])


def newline(p1, p2, color = 'black'):
    ax = plt.gca()
    xmin, xmax = ax.get_xbound()

    if(p2[0] == p1[0]):
        xmin = xmax = p1[0]
        ymin, ymax = ax.get_ybound()
    else:
        ymax = p1[1]+(p2[1]-p1[1])/(p2[0]-p1[0])*(xmax-p1[0])
        ymin = p1[1]+(p2[1]-p1[1])/(p2[0]-p1[0])*(xmin-p1[0])

    l = mlines.Line2D([xmin,xmax], [ymin,ymax], color = color)
    ax.add_line(l)
    return l

def check_and_draw(x, y, approximate_function, point):
    fig, ax = plt.subplots()   
    xnew=np.linspace(np.min(x),np.max(x),100)
    ynew=[approximate_function(x,y,i)[0] for i in xnew]
    plt.plot(x,y,'o', label = 'input data')
    plt.plot(xnew,ynew, label = 'approximate function')
    plt.plot( point[0], point[1], '*', markersize=12, label = 'approximation')
    plt.title(approximate_function)
    ax.legend() 
    plt.grid(True)
    plt.show()
```

#### Calculate interpolation

```python
### ЛАГРАНЖ
def lagrange(array_x, array_y, cur_x):
    array_x.astype(float)
    array_y.astype(float)
    lag = 0
    lagrangians = []
    for j in range(len(array_y)):
        multiplying = 1
        for i in range(len(array_x)):
            if i != j: 
                multiplying *= (cur_x - array_x[i]) / (array_x[j] - array_x[i])
        lagrangians.append(array_y[j] * multiplying)
        lag += array_y[j] * multiplying
    return lag, lagrangians

### НЬЮТОН
# посчитаем матрицу коэффициентов
def count_coef(array_x, array_y):
    m = len(array_x)
    
    array_x.astype(float)
    array_y.astype(float)
    array_x = np.copy(array_x)
    array_y = np.copy(array_y)
#     helb = np.copy(array_y)
#     need = [np.copy(array_y)]
    for k in range(1, m):
#         helb[k:m] = helb[k:m] - helb[k - 1:m-1]
#         need.append(np.copy(helb[k:m]))
        array_y[k:m] = (array_y[k:m] - array_y[k - 1])/(array_x[k:m] - array_x[k - 1])
#     return array_y, need
    return array_y


def newton_polynomial(array_x, array_y, cur_x):
#     coef, need = count_coef(array_x, array_y)
    coef = count_coef(array_x, array_y)

    n = len(array_x) - 1  # Degree of polynomial
    cur_y = coef[n]

    for k in range(1, n + 1):
        cur_y = coef[n - k] + (cur_x - array_x[n - k]) * cur_y
    
#     t = -0.7
#     ans = 0
#     for i in range(len(need)):
#         coeff = 1
#         for tt in range(0, i):
#             coeff *= t+tt
#         coeff /= fc(i)
#         print(need[i][-1]*coeff)
#         ans += need[i][-1]*coeff
        
    
    return cur_y, None

## НЬЮТОН где 1 и 2 порядок
def newton_simple(array_x, array_y, cur_x):
    array_x.astype(float)
    array_y.astype(float)
    full_array = np.array([array_x.astype(float), array_y.astype(float)]).transpose()
    full_array = full_array[np.argsort(full_array[:,0])].transpose()
    pivot = -1
    for i in range(1,len(array_x)):
        if full_array[0][i-1] < cur_x < full_array[0][i]:
            pivot = i-1
            break
    if pivot == 0:
        pass
    elif pivot == len(array_x) - 2:
        pivot -= 2
    else:
        pivot -=1
    f01 = (full_array[1][pivot+1]-full_array[1][pivot]) / (full_array[0][pivot+1]-full_array[0][pivot])
    f12 = (full_array[1][pivot+2]-full_array[1][pivot+1]) / (full_array[0][pivot+2]-full_array[0][pivot+1])
    f23 = (full_array[1][pivot+3]-full_array[1][pivot+2]) / (full_array[0][pivot+3]-full_array[0][pivot+2])
    f012 = (f12-f01)/ (full_array[0][pivot+2]-full_array[0][pivot])
    f123 = (f23-f12)/ (full_array[0][pivot+3]-full_array[0][pivot+1])
    cur_y_0 = full_array[1][pivot] + f01 * (cur_x - full_array[0][pivot]) + \
            f012 * (cur_x - full_array[0][pivot]) * (cur_x - full_array[0][pivot+1]) 
    cur_y_1 = full_array[1][pivot+1] + f12 * (cur_x - full_array[0][pivot+1]) + \
            f123 * (cur_x - full_array[0][pivot+1]) * (cur_x - full_array[0][pivot+2]) 
    cur_y = (cur_y_0 + cur_y_1)/2
    
    return cur_y, cur_y_0, cur_y_1
    
```

```python
def run():
    
    again = True
    print('Добрый день! Будем интерполировать!\n')
    while again:
        again = False
        in_type = input('Для начала нужны данные. Введи:\n\t* k - если хочешь вводить с клавиатуры\n\t* f - если хочешь вводить из файла\n')
        if in_type.strip() == 'k':
            data = input_vals()
        elif in_type.strip() == 'f':
            data = parse()
        else:
            print('Введено неверно, попробуй снова.')
            again = True
    print(f'Данные с которыми работаем:\n{data[0]}\n{data[1]}')

    cur_x = in_float('Введи число, для которого ты хочешь интерполировать значение: ', check = [True, min(data[0]), max(data[0])])
    
    lag, lagrangians = lagrange(data[0],data[1],cur_x)
#     print(f'\nУПРОЩЕННЫЙ ЛАГРАНЖ\nЧлены суммы при подсчете Лагранжианы: {lagrangians}')
    print(f'\nЛАГРАНЖ')
    print(f'Ответ методом Лагранжа: {lag}')
    check_and_draw(data[0], data[1], lagrange, [cur_x, lag])
    print()
    
    cur_y, cur_y_0, cur_y_1 = newton_simple(data[0],data[1],cur_x)
    print(f'\nНЬЮТОН (упрощенный)\nМетодом Ньютона, учитывая только разделенные разности 1 и 2 порядков ответы:'+
          f'\n{cur_y_0}\n{cur_y_1}\nУсредненный ответ: {cur_y}')
    check_and_draw(data[0], data[1], newton_simple, [cur_x, cur_y])
    print()
    

    newtone_answer = newton_polynomial(data[0],data[1],cur_x)[0]
    print(f'НЬЮТОН\nОтвет методом Ньютона: {newtone_answer}')
    check_and_draw(data[0], data[1], newton_polynomial, [cur_x, newtone_answer])
```
